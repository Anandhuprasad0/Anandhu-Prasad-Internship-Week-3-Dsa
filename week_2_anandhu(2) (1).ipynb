This code will winsorize each numeric column in the DataFrame, replacing extreme values with less extreme values determined by the specified limits.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Defining a function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Good', 'Standard', 'Poor'],
                yticklabels=['Good', 'Standard', 'Poor'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Plot confusion matrices for each model
plot_confusion_matrix(y_test, y_pred_lr, 'Logistic Regression')
plot_confusion_matrix(y_test, y_pred_dt, 'Decision Tree')
plot_confusion_matrix(y_test, y_pred_rf, 'Random Forest')
plot_confusion_matrix(y_test, y_pred_knn, 'K-Nearest Neighbors')
plot_confusion_matrix(y_test, y_pred_gb, 'Gradient Boosting')
plot_confusion_matrix(y_test, y_pred_nn, 'Neural Network')
```
